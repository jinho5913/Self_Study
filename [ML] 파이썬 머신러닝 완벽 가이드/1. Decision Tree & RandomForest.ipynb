# 1. 결정 트리(Desicion Tree)

## 1) 결정 트리란?

__`결정 트리`__는 ML 알고리즘 중 가장 직관적으로 이해하기 쉬운 알고리즘이다. 데이터에 있는 규칙을 통해 트리 기반의 분류 규칙을 만들어 예측을 한다. 따라서 데이터의 어떤 기준을 바탕으로 규칙을 만들어 내느냐에 따라 모델 성능이 결정된다고 해도 무방하다.  
  
__`결정 트리`__를 간단히 설명하면 데이터에 내재 되어있는 패턴을 변수의 조합으로 나타내는 예측/분류 모델을 '나무(Tree)'의 형태로 만드는 것이다.

- 장점 : 해석력이 높음, 직관적, 범용성  
- 단점 : 높은 변동성, 표본에 따라 트리구조가 달라짐(민감) -> __과적합 가능성__  
  
  

결정 트리는 매우 쉽고 유연하게 적용될 수 있는 알고리즘이다. 또한 데이터의 스케일링이나 정규화등의 Data Processing의 영향이 적은 편이다. 하지만, 성능을 높이기 위해서 복잡한 규칙 구조를 가져야 하며, 이로 인해 과적합(Overfitting)이 발생할 수 있다.  
  
이러한 단점은 오히려 앙상블에서는 장점으로 적용할 수 있다. `앙상블`은 매우 많은 여러 개의 모델을 결합해 확률적 보완과 오류가 발생한 부분에 대한 가중치를 계속 업데이트하면서 예측 성능을 향상시키는데, 결정트리가 좋은 학습기가 되기 때문이다.  

## 2) 결정 트리의 구조

* __노드__ : 트리의 기본 구성요소  
* __루트 노드(Root Node)__ : 트리의 최상의 노드  
* __규칙 노드(Decision Node)__ : 예측 값을 분류하기 위해 규칙 조건을 가지는 노드. 자식 노드가 있는 브랜치 노드(결정 노드)  
* __리프 노드(Leaf Node)__ : 규칙노드에 의해 분류된 최종 예측 값. 더 이상 자식 노드가 없는 노드(말단 노드)  
* __서브 트리(Sub Tree)__ : 새로운 규칙 조건마다 생성되는 트리(브랜치)  

만약 규칙 노드가 많다면 어떤 현상이 발생할까?  
  
주어진 데이터에서 분류를 세부적으로 할 수 있다는 장점이 있지만, 이는 곧 분류 결정 방식이 복잡하다는 것이기 때문에 과적합이 발생할 수 있다.  즉, 트리의 깊이(depth)가 깊어질수록 결정 트리의 예측 성능이 저하될 가능성이 높다.  
  
따라서 가능한 적은 결정 노드로 높은 예측 정확도를 가지려면 데이터를 분류할 때 최대한 많은 데이터 세트가 해당 분류에 속할 수 있도록 규칙 노드의 규칙이 정해져야 한다.

## 3) 결정 트리의 예측 방식

결정 트리는 최대한 __`균일한`__ 데이터 세트를 구성할 수 있도록 분할하는 것이 유리하다.  
  
결정 노드는 __정보 균일도가 높은 데이터 세트__를 먼저 선택할 수 있도록 규칙 조건을 생성한다. 정보 균일도가 높은 데이터 세트로 쪼개질 수 있도록 조건을 찾아 서브 데이터 세트를 만들고, 다시 이 서브 데이터 세트에서 균일도가 높은 자식 데이터 세트를 쪼개는 방식을 자식 트리로 내려가면서 가지를 치는 형식으로 알고리즘이 구현된다. 

이러한 정보의 균일도를 측정하는 대표적인 방법은 엔트로피를 이용한 __`정보 이득 지수`__ 와 __`지니 계수`__ 가 있다.  
  
* `정보 이득 지수(Information Gain)` : $ 1 - \text{entropy} $  
엔트로피 : 데이터 셋의 혼잡도
> 엔트로피가 높다 -> 데이터 불균형이 강하다.  
> 엔트로피가 낮다 -> 데이터가 상대적으로 균형적이다.  
> 결정 트리는 `Informatoin Gain`로 분할 기준을 정한다.  
> 즉, __`Information Gain`__ 이 높은 속성을 기준으로 분할한다.  
  
      
* `지니 계수` : 불평등 지수를 나타내는 계수 -> 불균형 상태를 나타내는 계수  
0 ~ 1 ==> 균형 ~ 불균형
> 지니 계수가 가장 낮은 속성을 기준으로 분할한다.  
> 지니 불순도(지니 계수와 반대되는 개념)가 높은 속성을 기준으로 분할한다고 이해하면 편함.

## 4) 결정 트리의 특징

- 장점 : 해석력이 높음, 직관적, 범용성  
- 단점 : 높은 변동성, 표본에 따라 트리구조가 달라짐(민감) -> __과적합 가능성__  
  
  

결정 트리는 매우 쉽고 유연하게 적용될 수 있는 알고리즘이다. 또한 정보의 균일도만 고려하면 되기 때문에 데이터의 스케일링이나 정규화등의 Data preprocessing의 영향이 적은 편이다. 
  


하지만, 성능을 높이기 위해서 복잡한 규칙 구조를 가져야 하며 Tree의 depth가 깊어지고, 이로 인해 과적합(Overfitting)이 발생할 수 있다.  
이러한 단점은 오히려 앙상블에서는 장점으로 적용할 수 있다. `앙상블`은 매우 많은 여러 개의 모델을 결합해 확률적 보완과 오류가 발생한 부분에 대한 가중치를 계속 업데이트하면서 예측 성능을 향상시키는데, 결정트리가 좋은 학습기가 되기 때문이다.  

## 5) 결정 트리 파라미터

- __`min_sample_split`__  : 노드를 분할하기 위한 최소한의 데이터 sample 수. 작게 설정할수록 분할되는 노드가 많아져서 과적합 가능성 증가.  
1로 설정할 경우, 분할되는 노드가 많아져서 과적합 가능성 발생 -> 과적합 제어용도로 사용. Defalut= 2  
  
  
- __`min_sample_leaf`__ : 말단 노드(Leaf Node)가 되기 위한 최소한의 데이터 sample 수. 과적합 제어 용도  
imbalanced 데이터의 경우 특정 클래스의 데이터가 극도로 작을 수 있으므로 이 경우는 작게 설정 필요  
  
  
- __`max_feature`__ : 최적의 분할을 위해 고려할 최대 피쳐 개수. int로 지정하면 대상 피쳐의 개수, float로 지정하면 전체 피쳐 중 대상 피쳐의 퍼센트  
  
  
- __`max_depth`__ : 트리의 최대 깊이 규정. None으로 설정 시 완벽하게 클래스 결정 값이 될 때 까지 깊이를 계속 키워 분할 or 노드가 가지는 데이터 개수가 `min_sample_split`보다 작아질 때가지 계속 깊이 증가.  
깊이가 깊어지면 `min_sample_split` 설정대로 최대 분할하여 과적합할 수 있으므로 적절한 값으로 제어 필요  
  
  
- __`max_leaf_nodes`__ : 말단 노드(Leaf Node)의 최대 개수

## 6) 결정 트리 모델의 시각화

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import numpy as np
%matplotlib inline

dt_clf = DecisionTreeClassifier(random_state = 156)

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.2, random_state = 11)

dt_clf.fit(X_train, y_train)
```

```python
from sklearn.tree import export_graphviz

# export_graphviz()의 호출 결과로 out_file로 지정된 tree.dot 파일을 생성함. 
export_graphviz(dt_clf, out_file="tree.dot", class_names=iris.target_names , \
feature_names = iris.feature_names, impurity=True, filled=True)
```

```python
import graphviz

# 위에서 생성된 tree.dot 파일을 Graphviz 읽어서 Jupyter Notebook상에서 시각화 
with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

```python
# feature importance 추출 
print("Feature importances:\n{0}".format(np.round(dt_clf.feature_importances_, 3)))

# feature별 importance 매핑
for name, value in zip(iris.feature_names , dt_clf.feature_importances_):
    print('{0} : {1:.3f}'.format(name, value))

# feature importance를 column 별로 시각화 하기 
sns.barplot(x=dt_clf.feature_importances_ , y=iris.feature_names)
```

## 7) 결정 트리 과적합(Overfitting)

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
%matplotlib inline

# 2차원 시각화를 위해서 feature는 2개, 결정값 클래스는 3가지 유형의 classification 샘플 데이터 생성
X_features, y_labels = make_classification(n_features = 2, n_redundant = 0, n_informative = 2,
                                          n_classes = 3, n_clusters_per_class = 1, random_state = 0)

plt.scatter(X_features[:, 0], X_features[:, 1], marker='o', c=y_labels, s=25, cmap='rainbow', edgecolor='k')
plt.title("3 Class values with 2 Features Sample data creation")
plt.show()
```

각 피처가 X, Y 축으로 나열된 2차원 그래프이며, 3개의 클래스 값 구분은 색깔로 되어있다.  
이를 기반으로 결정 트리에 대해 알아보자.

```python
import numpy as np

# Classifier의 Decision Boundary를 시각화하는 함수
def visualize_boundary(model, X, y):
    fig, ax = plt.subplots()
    
    # 학습 데이터 scatter plot으로 나타내기
    ax.scatter(X[:,0], X[:,1], c = y, s = 25, cmap = 'rainbow', edgecolor = 'k', 
              clim = (y.min(), y.max()), zorder = 3)
    ax.axis('tight')
    ax.axis('off')
    xlim_start, xlim_end = ax.get_xlim()
    ylim_start, ylim_end = ax.get_ylim()
    
    # 호출 파라미터로 들어온 training 데이터로 학습
    model.fit(X, y)
    
    # meshgrid 형태인 모든 좌표값으로 예측 수행
    xx, yy = np.meshgrid(np.linspace(xlim_start,xlim_end, num=200),np.linspace(ylim_start,ylim_end, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    
    # contourf() 를 이용하여 class boundary 를 visualization 수행. 
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='rainbow', clim=(y.min(), y.max()),
                           zorder=1)
```

* 생성 제약이 없는 결정 트리의 학습과 결정 경계 시각화

```python
from sklearn.tree import DecisionTreeClassifier

# 특정한 트리 생성 제약없는 결정 트리의 Decsion Boundary 시각화.
dt_clf = DecisionTreeClassifier().fit(X_features, y_labels)
visualize_boundary(dt_clf, X_features, y_labels)
```

일부 이상치(Outer) 데이터까지 분류하기 위해 분할이 자주 일어나서 결정 기준 경계가 매우 많아졌다.  
  
결정 트리의 기본 하이퍼 파라미터 설정은 리프 노드 안에 데이터가 모두 균일하거나 하나만 존재해야 하는 엄격한 분할 기준으로 인해 결정 기준 경계가 많아지고 복잡해졌다.  
  
이렇게 복잡한 모델은 학습 데이터 세트의 특성과 약간만 다른 형태의 데이터 세트를 예측하면 예측 정확도가 현저히 떨어진다. (비효율적)

* `min_sample_leaf = 6`으로 설정한 결정 트리 시각화

```python
# min_samples_leaf=6 으로 트리 생성 조건을 제약한 Decision Boundary 시각화
dt_clf = DecisionTreeClassifier( min_samples_leaf=6).fit(X_features, y_labels)
visualize_boundary(dt_clf, X_features, y_labels)
```

이상치에서 크게 반응하지 않으면서 더 일반화된 분류 규칙에 따라 분류가 되었다.  
  
다양한 테스트 데이터 세트를 기반으로 한 결정 트리 모델의 예측 성능은 **트리 생성 조건을 제약한 모델이 더 뛰어날 가능성이 높다**  
-> 테스트 데이터 세트는 학습 데이터 세트와는 다른 데이터 세트인데, 학습 데이터에만 지나치게 최적화된 분류 기준은 오히려 데이터 세트에서 정확도를 떨어뜨릴 수 있기 때문

# 2. 앙상블 학습(Ensemble Learning)

앙상블 학습(Ensemble Learning)은 여러 개의 분류기를 생성하고, 그 예측을 결합함으로써 보다 정확한 예측을 도출하는 기법을 말한다.  
강력한 하나의 모델을 사용하는 것 보다 약한 모델 여러개를 조합하여 더 정확한 예측에 도움을 주는 방식이다.  
  
ex) 하나의 문제에 대해서 한명의 전문가보다 여러 명의 전문가로 구성된 그룹이 보다 효율적이고 좋은 대안을 낼 수 있는 경우  
  
  
대부분의 정형 데이터 분류 시에는 앙상블이 뛰어난 성능을 나타내고 있다.  
`RandomForest`, `GradientBoost` 은 앙상블 알고리즘의 대표적인 모델들로 많은 사람들이 애용하고 있다.  
`GradientBoosting`을 뛰어넘는 `XGBoost`, `LightGBM`, `Stacking` 등이 개발되어 다양한 유형의 앙상블 알고리즘이 머신러닝의 선도 알고리즘으로 인기를 모으고 있다. 이는 분류, 회귀 분야에서 자주 쓰인다.  

앙상블 학습의 유형으로는 보팅(Voting), 배깅(Bagging), 부스팅(Boosting)이 있다.  

### 1) Voting(보팅)
  
서로 다른 여러 종류의 알고리즘을 사용한 각각의 결과에 대해 투표를 통해 최종 결과를 예측하는 방식  
`Soft Voting`과 `Hard Voting`으로 나뉜다.    

####  - Hard Voting 
  
예측한 결과값들 중 다수의 분류기가 결정한 예측값을 최종 보팅 결과값으로 선정하는 것 -> __`다수결`__과 비슷한 원리라고 생각하면 된다.  
  

다음 그림은 하드 보팅의 예시를 그림으로 나타낸 것이다. Test Data 셋에서 각각의 모델로 예측한 결과가 다음과 같다고 하자.  
__빨강 : 3 , 파랑 : 2__ 로 예측값이 나왔기 때문에 하드 보팅에서는 예측값이 가장 많은 __`빨강`__으로 최종 예측 결과값을 도출한다.

#### - Soft Voting 
  
분류기들의 타겟값 결정 확률을 모두 더하고 이를 평균해서 이들 중 확률이 가장 높은 타겟값을 최종 보팅 결과값으로 선정한다.  
  


다음 그림은 소프트 보팅의 예시를 그림으로 나타낸 것이다. Test Data 셋에서 각각의 모델로 예측한 레이블 값의 확률이 다음과 같다고 하자.  
각각의 확률을 평균한 결과, __빨강 : 0.41, 파랑 : 0.59__ 로 확률값이 나왔기 때문에, 소프트 보팅에서는 예측 레이블 값의 확률값이 가장 높은 __`파랑`__으로 최종 예측 결과값을 도출한다.

```python
# Hard Voting 레이블 별 확률값 게산
print('빨강 확률 :', round(((0.55+0.65+0.2+0.55+0.1)/5), 2))
print('파랑 확률 :', round(((0.45+0.35+0.8+0.45+0.9)/5), 2))
```

#### - Voting Classifier 실습

```python
import pandas as pd

from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

cancer = load_breast_cancer()

data_df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
data_df.head()
```

```python
# 개별 모델은 로지스틱 회귀와 KNN
lr_clf = LogisticRegression(solver = 'lbfgs', class_weight = 'balanced', max_iter = 10000)
knn_clf = KNeighborsClassifier(n_neighbors = 8)
rf_clf = RandomForestClassifier(random_state = 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                   test_size = 0.2, random_state = 156)
```

```python
# 개별 모델을 소프트 보팅 기반의 앙상블 모델로 구현한 분류기
vo_soft_clf = VotingClassifier(estimators = [('LR',lr_clf),('KNN',knn_clf), ('RF', rf_clf)], voting = 'soft')

# 소프트 보팅 모델 학습
vo_soft_clf.fit(X_train, y_train)
pred_s = vo_soft_clf.predict(X_test)
```

```python
# 개별 모델을 하드 보팅 기반의 앙상블 모델로 구현한 분류기
vo_hard_clf = VotingClassifier(estimators = [('LR',lr_clf),('KNN',knn_clf), ('RF', rf_clf)], voting = 'hard')

# 하드보팅 모델 학습
vo_hard_clf.fit(X_train, y_train)
pred_h = vo_hard_clf.predict(X_test)
```

```python
print('Soft Voting 분류기 정확도: ', accuracy_score(y_test, pred_s))
print('Harrd Voting 분류기 정확도: ', accuracy_score(y_test, pred_h))

# 개별 모델의 학습/예측/평가
classifier = [lr_clf, knn_clf, rf_clf]
for classifier in classifier:
    classifier.fit(X_train, y_train)
    pred = classifier.predict(X_test)
    class_name = classifier.__class__.__name__
    print('{0} 정확도: {1:.4f}'.format(class_name, accuracy_score(y_test , pred)))
```

`보팅 분류기 정확도 > 개별 모델 정확도` 임을 알 수 있다.  
  
하지만, 항상 앙상블 기법이 높은 성능을 보이는건 아니다.  
  
또한, Hard Voting 보다 Soft Voting이 일반적으로 성능 결과가 더 잘 나오는 편이라 합리적이다.

### 2) Bagging(배깅)  
  
같은 알고리즘에 대해 서로 다른 데이터 셋을 학습시킨 후, 투표를 통해 예측값을 구하는 방법  
대표적인 배깅 방식이 `RandomForest`  
  
__Bootstrap__(Random Sampling) + __Aggregating__(집계) = __Bagging__

다음 그림은 Bagging의 파이프라인을 그림으로 나타낸 것이다.  

단일 ML 알고리즘(Decision Tree)으로 여러 분류기가 학습으로 개별 예측을 하는데, 학습하는 데이터 세트가 보팅 방식과는 다름을 알 수 있다.  
  

개별 분류기에 할당된 학습 데이터는 원본 학습 데이터를 샘플링해 추출하는데, 이렇게 개별 Classifier에게 데이터를 샘플링해서 추출하는 `방식을 부트스트래핑(Bootstraping) 분할 방식`이라 한다.  
개별 분류기가 부트스트래핑 방식으로 샘플링된 데이터 세트에 대해서 학습을 통해 개별적인 예측을 수행한 결과를 Voting을 통해서 최종 예측 결과를 선정하는 방식이 Bagging이다.

#### - __Bootstraping__  
  
데이터를 랜덤샘플링하는 방법으로, 복원추출을 한다.  
이론적으로 한 개체가 하나의 bootstrap에 한번도 선택되지 않을 확률은 다음과 같다. 
  
  

  
  
주어진 데이터가 충분하지 않을 때, 임의적으로 데이터의 양을 늘린다.  
데이터를 약간 편향되도록 샘플링하는 기법으로, 샘플링을 통해 편향을 높여서 분산이 높은 모델의 과적합을 줄이는 효과를 제공한다.  
  
앞서 소개했던 Decision Tree는 __high variance__로 인한 약점을 가지고 있었다.(분산이 높다는 것은 다른 데이터셋에 따라 모델이 심하게 변동한다는 것을 의미) Bagging은 이런 __Variance__를 줄여주기 위한 목적을 가지고 있다. 특히 Decision Tree에서 효율적으로 쓰인다.  
  
bootstrap을 통해 N 개의 training set을 만들고, 이에 대해 N 개의 tree를 만드는것이다. 이때, tree들은 제거하지 않는다. 따라서 low bias이지만, 매우 높은 variance를 가지고 있는 모델이다. 이들을 최종적으로 평균내줌으로써, variance를 줄여주는 것이다.  
  
물론 Bootstrap을 통해 만들어진 data set들은 그 구성도 매우 비슷하여, 모델들간의 공분산도 높아진다!

#### - __Out-of-Bag Error__  
  
Bagging을 통해서는 각 Bootstrap을 하며 복원추출에 뽑히지 않은 데이터들(보통 2/3 정도가 뽑힌다)이 자동으로 validation set이 되어 cv를 하지 않아도 error값을 추정할 수 있다. 여기서 뽑히지 않은 데이터들을 __`Out-of-Bag(OOB)`__라고 한다.  
  
특정 데이터가 뽑히지 않은, 전체 N 개 중 약 N/3개의 모델들이 특정 데이터에 대한 예측을 하고 이를 평균낸다. 이런 식으로 모든 데이터에 대해 예측을 하여 OOB MSE나, Classifier error를 구할 수 있다. 물론 cv가 더 정확한 error 추정을 가능하게 하지만, 데이터가 충분히 많고 충분히 많은 N번의 Bootstarp의 경우, OOB 역시 비슷한 성능을 내기 때문에, 별도의 cv를 진행하지 않는다는 점에서 차이점이 있다.  
___-> 그래도 cv를 하자..!___

#### - __Aggregating__  
  
여러 분류 모델이 예측한 값들을 조합해서 하나의 결론을 도출하는 과정이다.  
연속형 변수의 경우 평균을, 범주형 변수의 경우 다수 투표로 집계를 한다.

# 3. 랜덤 포레스트(Random Forest)

## 1) 랜덤 포레스트란?

배깅은 보팅과 다르게 같은 알고리즘으로 여러 개의 분류기를 만들어서 보팅으로 최종 결정하는 알고리즘이다.  
배깅의 대표적인 알고리즘은 `랜덤 포레스트(Random Forest)`이다. Decision Tree의 단점인 과적합을 회피할 목적으로 Decision Tree를 base 모델로 하는 앙상블 모델로, 다수의 Decision Tree 모델에 의한 예측/분류값을 종합하는 방법이다. 물론 결정 트리의 쉽고 직관적인 장점을 그대로 가지고 있다.  
  
-> 랜덤 포레스트뿐만 아니라, 부스팅 기반의 다양한 앙상블 알고리즘 역시 대부분 결정 트리 알고리즘을 기반으로 알고리즘을 채택하고 있다.)  
  
  

랜덤 포레스트는 개별적인 분류기의 기반 알고리즘은 결정 트리이지만, 개별 트리가 학습하는 Data Set는 전체 데이터에서 일부가 중첩되게 샘플링된 Data Set 이다.(Bootstraping 분할 방식으로 샘플링되었기 때문이다.)  

랜덤 포레스트의 서브세트(Subset) 데이터는 이러한 부트스트래핑으로 데이터가 임의로 만들어진다. 서브세트의 데이터 건수는 전체 데이터 건수와 동일하지만, 개별 데이터가 중첩되어 만들어진다. 원본 데이터의 건수가 10개인 학습 데이터 세트에 랜덤 포레스트를 n개의 결정 트리 기반으로 학습하려고 `n_estimators = n`으로 설정한다면, 다음과 같이 데이터 서브세트가 만들어진다.  
  


랜덤 포레스트의 핵심 keyword는 __무작위성 "random"__ & __다양성 "diversity"__이다.  
  
`무작위성`은 트리를 더욱 다양하게 만들고, `다양성`은 분산을 낮추어 더 훌륭한 모델을 만들어낸다.  
  

__< 랜덤 포레스트 단계 >__  
1. 부트스트랩을 통한 학습 데이터 샘플링 (다양성)  

2. 개별 Decision Tree 훈련 (무작위성)-> 전체 피쳐 개수가 m개라면 $\sqrt{m}\$개를 랜덤으로 뽑아 피쳐로 사용

3. 개별 모델을 통한 예측  

4. 결과 도출  
  
개별 Tree의 성능이 좋지 않아도, 다양성이 확보된 전체 모델의 성능은 향상되는 경우가 빈번하다.  
  
__< 랜덤 포레스트의 장점 >__  
1. Classification(분류) & Regression(회귀) 문제에 모두 사용 가능  
2. Missing Value(결측치)를 다루기 쉬움  
3. 모델의 노이즈를 심화시키는 OverFitting 문제를 회피하며, 모델 정확도를 향상시킴  
4. Classification 모델에서 상대적으로 중요한 변수 시각화 가능

## 2) 랜덤포레스트 하이퍼 파라미터

Decision Tree와 거의 동일한 하이퍼 파라미터를 가지고 다음은 조금 다른 파라미터들이다.  
  
* __`n_estimators`__ : 결정트리의 갯수를 지정, 무작정 트리 갯수를 늘리면 성능이 좋아지는 것 대비 시간이 걸릴 수 있음(Default = 10)  
  
  
* __`max_features`__ : 최적의 분할을 위해 고려할 최대 feature의 개수  
결정 트리 : 노드를 분할할 때 모든 feature를 고려 : default = None  
랜덤포레스트 : 랜덤으로 선택 된 feature를 고려 : default =  'auto' -> $\sqrt{m}\$

## 3) RandomForestClassifier 실습

### - RandomForest

```python
import pandas as pd

def get_human_dataset():
    
    # 각 데이터 파일들은 공백으로 분리되어 있으므로 read_csv에서 공백 문자를 sep으로 할당.
    feature_name_df = pd.read_csv('data_set/human_activity/features.txt',sep='\s+',
                        header=None,names=['column_index','column_name'])
    # DataFrame에 피처명을 컬럼으로 부여하기 위해 리스트 객체로 다시 변환
    feature_name = feature_name_df.iloc[:, 1].values.tolist()
    
    # 학습 피처 데이터 셋과 테스트 피처 데이터을 DataFrame으로 로딩. 컬럼명은 feature_name 적용
    X_train = pd.read_csv('data_set/human_activity/train/X_train.txt',sep='\s+')
    X_train.columns = feature_name
    X_test = pd.read_csv('data_set/human_activity/test/X_test.txt',sep='\s+')
    X_test.columns = feature_name
    
    # 학습 레이블과 테스트 레이블 데이터을 DataFrame으로 로딩하고 컬럼명은 action으로 부여
    y_train = pd.read_csv('data_set/human_activity/train/y_train.txt',sep='\s+',names=['action'])
    y_test = pd.read_csv('data_set/human_activity/test/y_test.txt',sep='\s+',header=None,names=['action'])
    
    # 로드된 학습/테스트용 DataFrame을 모두 반환 
    return X_train, X_test, y_train, y_test
```

```python
from sklearn.metrics import accuracy_score
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# 결정 트리에서 사용한 get_human_dataset( )을 이용해 학습/테스트용 DataFrame 반환
X_train, X_test, y_train, y_test = get_human_dataset()

# 랜덤 포레스트 학습 및 별도의 테스트 셋으로 예측 성능 평가
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train , y_train)
pred = rf_clf.predict(X_test)
accuracy = accuracy_score(y_test , pred)
print('랜덤 포레스트 정확도: {0:.4f}'.format(accuracy))
```

### - RandomForest 하이퍼 파라미터 및 튜닝
  
트리 기반의 앙상블 알고리즘의 단점은 하이퍼 파라미터가 너무 많고, 그로 인해서 튜닝을 위한 시간이 많이 필요하다는 것이다.

```python
# GridSearchCV를 이용한 랜덤 포레스트의 하이퍼 파라미터 튜닝

from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators':[100],
    'max_depth' : [6, 8, 10, 12], 
    'min_samples_leaf' : [8, 12, 18 ],
    'min_samples_split' : [8, 16, 20]
}

rf_clf = RandomForestClassifier(random_state = 0, n_jobs = -1) # n_jobs = -1 이용시 모든 CPU 코어 사용
grid_cv = GridSearchCV(rf_clf, param_grid = params, cv = 2, n_jobs = -1)
grid_cv.fit(X_train, y_train)

print('최적 하이퍼 파라미터:\n', grid_cv.best_params_)
print('최고 예측 정확도: {0:.4f}'.format(grid_cv.best_score_))
```

```python
# n_estimator를 300으로 증가 시키고 별도의 테스트 데이터 세트에서 예측 성능 특정 진행
rf_clf_1 = RandomForestClassifier(n_estimators = 300, max_depth = 10, min_samples_leaf = 12,
                                 min_samples_split = 8, random_state = 0)
rf_clf_1.fit(X_train, y_train)
pred = rf_clf_1.predict(X_test)
print('예측 정확도 : {0:.4f}'.format(accuracy_score(y_test, pred)))
```
